```python
def normalize(image_set):
  normalized = [img/255 for img in image_set]
  return normalized
```

```python
normalised_image_array = []
folders = os.listdir("/content/drive/MyDrive/ImageProcessing/")
for folder in folders:
  for path in os.listdir("/content/drive/MyDrive/ImageProcessing/"+folder):
    paths = os.path.join("/content/drive/MyDrive/ImageProcessing/"+folder+"/"+path)
    image = Image.open(paths)
    image = np.asarray(image)
    normalized_image =  normalize(image)
    normalised_image_array.append(normalized_image)
```
